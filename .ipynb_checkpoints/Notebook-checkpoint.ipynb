## Memulai dengan read data file
dimulai dengan import pandas sebagai feb dan os kemudian read pokemon_data di dalam folder folder data
dilanjutkan dengan menampilkan data head 10baris

```python
# Import pandas sebagai feb
import pandas as feb
import os
import seaborn as feb1
from matplotlib import pyplot as feb2

# Raed data pokemon_data.csv didalam folder data
fa = feb.read_csv('data/pokemon_data.csv')

# Menampilkan data dalam 10 baris
fa.head(10)
```

## Reading Data di Pandas
mengetahui terdapat index apa saja didalam data

```python
# Read Headers
fa.columns
```

```python
# Read tiap kolom
print(fa[['Name', 'Type 1', 'Type 2']])
```

# # Mendeskripsikan dengan lengkap
menggunakan describe()

```python
fa.describe()
```

# # Sorting data
menggunakan sort_values untuk sorting index Type 1 dan HP seacara double dan ascending true agar dimulai dengan sort paling sedikit

```python
# Menyorting Data

fa.sort_values(['Type 1', 'HP'], ascending=True)

fa.head(10)
```

## Menghitung total data
menghitung total data dengan menjumlahkan HP, Attack, Defense, Sp. Atk, SP. Def dan Speed dan disimpan kedalam kolom Total

```python
fa['Total'] = fa['HP'] + fa['Attack'] + fa['Defense'] + fa['Sp. Atk'] + fa['Sp. Def'] + fa['Speed']

fa.head(10)
```

```python
45+49+49+65+65+45
```

## Menghitung total dari HP, Attack, Defense, Sp. Atk, Sp. Def dan Speed

```python
# Menghitung total HP
fa.loc['Total', 'HP'] = fa['HP'].sum()

# Menghitung total Attack
fa.loc['Total', 'Attack'] = fa['Attack'].sum()

# Menghitung total Defense
fa.loc['Total', 'Defense'] = fa['Defense'].sum()

# Menghitung total Sp. Atk
fa.loc['Total', 'Sp. Atk'] = fa['Sp. Atk'].sum()

# Menghitung total Sp. Def
fa.loc['Total', 'Sp. Def'] = fa['Sp. Def'].sum()

# Menghitung total Speed
fa.loc['Total', 'Speed'] = fa['Speed'].sum()
fa
```

```python
fa
```

## Hapus kolom total

```python
fa = fa.drop(columns=['Total'])
```

```python
fa
```

## Menampilkan bar
Bar plotting jumlah pokemon berdasarkan type 1 dan type 2 menggunakan module seaborn. dan module ptplot dari matplotlib untuk mengubah rotasi x jadi -45 agar tulisan terbaca

```python
# bar type 1
feb1.countplot(x='Type 1', data=fa)
feb2.xticks(rotation=-45)
```

```python
# bar type 2
feb1.countplot(x='Type 2', data=fa)
feb2.xticks(rotation=-45)
```

## Menyimpan data sekarang
menyimpan data kedalam folder data. 

```python
# Menghapus hasil sebelummnya agar bisa menyimpan file baru
#Simpan file sebagai CSV
fa.to_csv("data/Modifiedcsv.csv", index = False)
#Simpan file sebagai XLSX
fa.to_excel("data/Modified.xlsx", index = False)
#Simpan file sebagai TXT
fa.to_csv("data/Modified.txt", index = False, sep = '/')
```

## Memfilter Data dan menyimpannya

```python
# Memfilter data type 1 adalah grass dan type 2 adalah poison
new_fa = fa.loc[(fa['Type 1'] == 'Grass') & (fa['Type 2'] == 'Poison')] 
new_fa.reset_index(drop=True, inplace=True)

new_fa

# mengurutkan nama dari a-z
new_fa.sort_values(['Name'])

# file disimpan dalam folder data dengan nama file filtered.csv

new_fa.to_csv('data/filtered.csv')
```
